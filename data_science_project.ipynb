```python
cities = data['Name'].unique().tolist()
len(cities)
```

```python

#creating list of tuples with the cities names
city_names = []
for city in cities:
    #removing unnecessary strings from the name of the city.
    if city == 'Anchorage municipality, Alaska':
        city_names.append(('Anchorage, Alaska',city))
        continue
    if city == 'Louisville/Jefferson County metro government (balance), Kentucky':
        continue
    if city == 'Urban Honolulu CDP, Hawaii':
        city_names.append(('Honolulu, Hawaii',city))
        continue
    pattern = re.search(r"(.*) \(.*\)(.*)",city)
    if pattern:
        name = pattern.group(1) + pattern.group(2)
        city_names.append((name.replace(" city",""),city))
        continue
    if city.find("CDP") >= 0:
        city_names.append((city.replace(" CDP",""),city))
        continue
    else:
        city_names.append((city.replace(" city",""),city))
    
city_names
        
```

```python
#deviding the names to chuncks for later use
chunk_size = 10
city_name_copy = city_names
chuncks = []
while city_name_copy:
    chunk, city_name_copy = city_name_copy[:chunk_size], city_name_copy[chunk_size:]
    chuncks.append(chunk)
chuncks
```
